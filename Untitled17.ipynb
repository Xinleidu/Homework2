```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ride_csv_path="/Users/lauradu/Desktop/PythonExercise/ride_data.csv"
```

```python
ride_df = pd.read_csv(ride_csv_path)
ride_df.head()
```

```python
city_csv_path="/Users/lauradu/Desktop/PythonExercise/city_data.csv"
```

```python
city_df=pd.read_csv(city_csv_path)
city_df.head()
```

```python
df1 = city_df.groupby(['city','type']).agg({'driver_count':np.sum}).reset_index()
df2 = ride_df.groupby('city').agg({'fare':np.mean}).reset_index()
df3 = ride_df.groupby('city').agg({'ride_id':'count'}).reset_index()

df = pd.merge(df1,df2,how='outer',on='city')
df = pd.merge(df,df3,how='outer',on='city')
```

```python
df.head()
```

```python
plt.scatter(x=df[df['type'] == 'Urban']['ride_id'], y=df[df['type'] == 'Urban']['fare'], c='Lightcoral', s = df[df['type'] == 'Urban']['driver_count']**1.5,
alpha = 0.75,label = 'Urban',edgecolors='black',linewidths=0.75)

plt.scatter( x=df[df['type'] == 'Suburban']['ride_id'],y=df[df['type'] == 'Suburban']['fare'],c= 'LightSkyBlue',s = df[df['type'] == 'Suburban']['driver_count']**1.5,
alpha = 0.60,label = 'Suburban',edgecolors="black",linewidths=0.75)

plt.scatter(x=df[df['type'] == 'Rural']['ride_id'],y=df[df['type'] == 'Rural']['fare'],c= 'Gold',s = df[df['type'] == 'Rural']['driver_count']**1.5,
alpha = 0.60,label = 'Rural',edgecolors="black",linewidths=0.75)

plt.legend( title="City type")
plt.xlabel("Total Number of rides (per city)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber ride sharing data")
```

```python
pie_fare = df.groupby('type').agg({'fare':np.sum}).reset_index().sort_values('type',ascending=False)
pie_fare
```

```python
labels = pie_fare['type']
sizes = pie_fare['fare']
colors = ['Gold', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0.05) 
plt.pie(sizes, colors=colors,explode=explode, autopct='%1.1f%%',labels=labels,startangle=48,
       wedgeprops   = { 'linewidth' : 0.5,'edgecolor' : "black" },shadow=True,
       labeldistance=1.10)
plt.axis('equal')
plt.title("Total Fares by City Type")
plt.legend()
plt.show()
```

```python
pie_rides = df.groupby('type').agg({'ride_id':np.sum}).reset_index().sort_values('type',ascending=False)
pie_rides
```

```python
labels = pie_rides['type']
sizes = pie_rides['ride_id']
colors = ['Gold', 'lightcoral', 'lightskyblue']
explode = (0, 0.10, 0.05) 
plt.pie(sizes, colors=colors,explode=explode, autopct='%1.1f%%',labels=labels,startangle=47,
       wedgeprops   = { 'linewidth' : 0.5,'edgecolor' : "black" },shadow=True,
       labeldistance=1.10)
plt.axis('equal')
plt.title("Total Rides by City Type")
plt.legend()
plt.show()
```
